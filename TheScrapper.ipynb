```python
from urllib.request import Request , urlopen  
from bs4 import BeautifulSoup as soup 



my_url = 'https://www.carrefour.es/monitores/cat410423/c'

req =  Request(my_url,headers={'User-Agent': 'Mozilla/5.0'})

# opening up connection , grabbing the page 

uClient =  urlopen(req)

page_html =  uClient.read()

# closes connection 

uClient.close()

# html parsing 

page_soup = soup(page_html,'html.parser')

containers = page_soup.find_all("div", class_="product-card")


contain = containers[0]

```

```python
def get_set(my_url):
    req =  Request(my_url,headers={'User-Agent': 'Mozilla/5.0'})
    uClient =  urlopen(req)
    page_html =  uClient.read()
    uClient.close()
    page_soup = soup(page_html,'html.parser')
    return page_soup
```

# First attempt 

```python
list_names = list()
list_prices = list()
for i in range(len(containers)):
    list_names.append(containers[i].a.img['alt'])
    list_prices.append(containers[i].find("span","product-card__price"))
    if list_prices[i] == None: 
        list_prices[i] = "No value 0  "
    else: list_prices[i] = list_prices[i].text
    list_prices[i] = int("".join(filter(str.isdigit, list_prices[i])))
    if list_prices[i] >= 1000:
        list_prices[i] = list_prices[i]/100
    
```

```python
import pandas as pd 

dataframe = pd.DataFrame(
    {'Product Name ': list_names,
     'Price': list_prices
    })
```
